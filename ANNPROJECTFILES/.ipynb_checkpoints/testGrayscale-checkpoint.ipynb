```python
import cv2
import glob
import os

# Define the path to the folder containing the color images
input_folder = '/Users/yashp/OneDrive/Desktop/Project/test/good/'

# Define the path to the folder where the grayscale images will be saved
output_folder = '/Users/yashp/OneDrive/Desktop/Project/test/Combinedtest/'

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Loop through all images in the input folder
for filename in glob.glob(os.path.join(input_folder, '*.jpg')):
    # Load the color image
    img_color = cv2.imread(filename)

    # Convert the color image to grayscale
    img_gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)

    # Save the grayscale image
    output_filename = os.path.join(output_folder, os.path.basename(filename))
    cv2.imwrite(output_filename, img_gray)
```
